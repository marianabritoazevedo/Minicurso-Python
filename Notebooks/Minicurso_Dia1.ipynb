O que vimos anteriormente foram as **express√µes boolenas**. J√° as **opera√ß√µes boolenas** podem envolver 3 tipos de opera√ß√µes:

*   **OU**: apenas uma express√£o precisa ser verdadeira para o resultado final ser verdadeiro. Na programa√ß√£o, usamos a palavra `or` para esta opera√ß√£o.
*   **NOT**: se algo √© verdade, se tornar√° falso, e se algo √© falso, se tornar√° verdade. Usamos a express√£o `not` na programa√ß√£o para esta opera√ß√£o.
