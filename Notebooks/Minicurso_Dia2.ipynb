## Parte 3: Estruturas de repeti√ß√£o

Em muitas situa√ß√µes, desejamos fazer uma mesma a√ß√£o v√°rias vezes em nossos programas. Para fazer um programa repetir um mesmo c√≥digo v√°rias vezes, podemos utlizar duas estruturas de repeti√ß√£o: o **for** (do ingl√™s, para) e o **while** (do ingl√™s, enquanto). Observe os exemplos abaixo para entender melhor como esse tipo de estrutura funciona.

### **Exemplo 3**: Contando de 1 a 10

O programa abaixo ir√° imprimir uma mensagem contando os n√∫meros de 1 a 10. Perceba que, para isso, utilizamos a estrutura de repeti√ß√£o `for`, e em seguida, uma estrutura `range(1, 11)`. Essa √∫ltima estrutura significa que vamos contar os n√∫meros dentro de uma certa `condi√ß√£o`, em que essa condi√ß√£o √© justamente que os n√∫meros sejam justamente entre 1 e 10, nem maiores nem menores do que isso. 

**E por que utilizamos os valores 0 a 11?**

No Python, ele ir√° contar at√© o √∫ltimo valor - 1, ent√£o, para que ele consiga contar at√© 10, √© preciso adicionar 1 em seu valor final para ele captar esse n√∫mero. Se tiv√©ssemos colocado `range(1, 10)`, ele s√≥ iria contar at√© o 9.

````python
for numero in range(0, 11):
  print('N√∫mero ' + str(numero))
````

### **Exemplo 4**: Acertando o valor da senha

Neste exemplo, a estrutura de repeti√ß√£o `while` ser√° utilizada. Ela √© utilizada quando n√£o sabemos exatamente quantas vezes queremos repetir o nosso programa. No exemplo anterior, sab√≠amos que quer√≠amos repetir nosso programa exatamente 10 vezes, para contar de 1 a 10, mas para adivinhar o valor de uma senha, n√£o sabemos quantas vezes o usu√°rio ir√° errar at√© finalmente acert√°-la. Vejamos o exemplo abaixo para adivinhar uma senha:

````python
senha_correta = "senha1234"
senha_digitada = input("Digite sua senha: ")

while(senha_correta != senha_digitada):
  print("A senha digitada est√° incorreta, tente novamente")
  senha_digitada = input("Digite sua senha: ")

print("Senha correta digitada!")
````

Enquanto o usu√°rio digitar a senha errada, ele n√£o ir√° avan√ßar no programa, ficar√° pedindo para o usu√°rio digitar a senha novamente. Por√©m, no momento que o usu√°rio digitar `senha1234`, nosso programa ver√° que a senha est√° correta, e assim, ir√° dizer que agora a senha digitada est√° correta

### **Exemplo 6**: Plotando um gr√°fico de barras com os dados do IBGE

A primeira parte, que j√° foi feita anteriormente e s√≥ ser√° repetida, √© justamente realizar a leitura dos dados do Excel e organiz√°-los em uma tabela. Em seguida, precisamos pegar esses dados, e utilizando a biblioteca Plotly, montar o gr√°fico desejado, que no nosso caso, √© um gr√°fico de barras. 

Dentro da funcionalidade de montar um gr√°fico de barras, temos que informar quais s√£o os dados que vamos querer plotar (ou seja, colocar nossa tabela inteira), e depois especificar quais valores ir√£o ficar no eixo x e no eixo y. 

````python
#Importando a biblioteca do pandas e do plotly
import pandas as pd
import plotly.express as px

#Criando a tabela com os dados do IBGE
dados = pd.read_excel('dados_ibge.xlsx')

#Criando a figura do gr√°fico a partir da tabela
fig = px.bar(dados, x='Ano', y='Popula√ß√£o')

#Mostrando a figura na tela
fig.show()
````

### **Exemplo 7**: Melhorando nosso gr√°fico

Nosso gr√°fico foi plotado, entretanto, tem certas coisas que poderiam ser melhoradas... Ele n√£o possui um t√≠tulo, e s√≥ consigo ver os valores exatos da popula√ß√£o ao colocar meu mouse em cima da barra. Tamb√©m queria poder mudar as cores da barra. Com o plotly, tudo isso √© poss√≠vel!

Para isso, precisamos apenas especificar esses valores com algumas funcionalidades a mais, veja no exemplo abaixo. O que mudou? Entre algumas mudan√ßas, temos:

*   **title**: do ingl√™s, t√≠tulo, ou seja, ser√° o t√≠tulo do nosso gr√°fico 
*   **color_discrete_sequence**: do ingl√™s, sequ√™ncia discreta de cores. Pode parecer um nome um pouco confuso, mas √© aqui que vamos definir a cor das barras do nosso gr√°fico. Podemos escrever o nome da cor em ingl√™s, ou utilizar um c√≥digo hexadecimal de cores. 
*   **text**: do ingl√™s, texto, ou seja, ir√° mostrar um texto em cada uma das barras do nosso gr√°fico. No nosso caso, queremos mostrar o texto com a popula√ß√£o brasileira em cada um dos anos.
*   **text_position=outside**: do ingl√™s, posi√ß√£o do texto = fora. Isso √© apenas um recurso de estiliza√ß√£o para dizer que queremos colocar o texto fora da barra. 



````python
#Importando a biblioteca do pandas e do plotly
import pandas as pd
import plotly.express as px

#Criando a tabela com os dados do IBGE
dados = pd.read_excel('dados_ibge.xlsx')

#Criando a figura do gr√°fico a partir da tabela
fig = px.bar(dados, x='Ano', y='Popula√ß√£o', title='Evolu√ß√£o da popula√ß√£o brasileira conforme dados do IBGE',
             color_discrete_sequence=['#586F94'], text='Popula√ß√£o')

#Mostrando a figura na tela
fig.update_traces(textposition='outside')
fig.show()
````
